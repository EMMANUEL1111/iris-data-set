# Confidence interval

```python
np.mean(data)
```

```python
import scipy.stats as st
```

```python
st.t.interval(0.99,len(data['sepal length (cm)'])-1,np.mean(data['sepal length (cm)']),np.std(data['sepal length (cm)']))
```

```python
st.t.interval(0.99,len(data['sepal width (cm)'])-1,np.mean(data['sepal width (cm)']),np.std(data['sepal width (cm)']))
```

```python
st.t.interval(0.99,len(data['petal length (cm)'])-1,np.mean(data['petal length (cm)']),np.std(data['petal length (cm)']))
```

```python
st.t.interval(0.99,len(data['petal width (cm)'])-1,np.mean(data['petal width (cm)']),np.std(data['petal width (cm)']))
```

```python
interval=st.t.interval(0.90,len(data)-1,np.mean(data),np.std(data))
```

```python
interval
```

```python
c_interval=pd.DataFrame(interval,index=['lower_bound','upper_bound'],columns=data.columns).T
```

```python
c_interval
```

```python
sns.regplot(x=c_interval['lower_bound'],y=c_interval['upper_bound'],data=c_interval)
```

# t-tests

```python
st.ttest_ind(data['sepal length (cm)'],data['sepal width (cm)'],axis=0,nan_policy='propagate')
```

```python
st.ttest_ind(data['petal length (cm)'],data['petal width (cm)'],axis=0,nan_policy='propagate')
```

# Z-score

```python
z_score=pd.DataFrame(st.stats.zscore(data,axis=0,ddof=0,nan_policy='propagate'),columns=data.columns)
z_score
```

```python
pd.DataFrame(st.stats.chisquare(data,axis=0),index=['observed ','expected'],columns=data.columns)
```

# Linear Regreesion

```python
X=data
y=iris.target
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=101)
from sklearn.linear_model import LinearRegression
random=LinearRegression(n_jobs=1)
random.fit(X_train,y_train)
predict=random.predict(X_test)
from sklearn.metrics import mean_absolute_error, mean_squared_error,explained_variance_score
```

```python
mean_absolute_error(y_test,predict)
```

```python
mean_squared_error(y_test,predict)
```

```python
np.sqrt(mean_squared_error(y_test,predict))
```

```python
explained_variance_score(y_test,predict)
```
